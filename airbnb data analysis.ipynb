```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

```python
data=pd.read_csv("compressed_data.csv")
```

Checking missing values

```python
data.isnull().sum()
```

Handling missing values

```python
data['last review']= pd.to_datetime(data['last review'],errors='coerce')
#errors='coerce' forces invalid values (like strings or junk data) to become NaN (Not a Number) instead of throwing an error.
```

```python
data.info()
```

```python
data.fillna({'reviews per month' : 0,'last review' : data['last review'].min()},inplace=True) #last reviwe me khali column ko minium date se fill kr denge 
#and reviwes per month wale columnme null ko 0 se eplce kr demnge because vo number wala colum .
```

```python
data.dropna(subset=['host name','NAME'],inplace=True)
```

```python
#inplace=true ka mtlb h data me dirct chnage bhi hoga jo hmnr load kra h
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['license','house_rules'],errors='ignore')
```

```python
data.head()
```

```python
data['price']=data['price'].replace('[\$,]','',regex=True).astype(float)
data['service fee']=data['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

Visualization

What is the distribution of the listing Prices?

```python
plt.figure(figsize=(8,6))
sns.histplot(data['price'],bins= 50,kde= True, color='blue')
plt.title("Distribution of Listing Prices")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
#kde yha pe curve bnane ke liye use hua h  taki pta chale ske ki kitni growth h
```

How are different room type distributed

```python
data['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data= data,color="hotpink")
plt.title(" Different room type distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="neighbourhood group",data= data,color="lightgreen",order=data['neighbourhood group'].value_counts().index)
plt.title(" No. of Listing by neighbourhood Group")
plt.xlabel("Count")
plt.ylabel('Neighbourhood group')
plt.show()
```

```python
#What is the relationship between price and room type
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='room type',y='price',hue='room type',data=data,palette='Set1')
plt.title('Price vs Room type')
plt.xlabel('room type')
plt.ylabel('Price($)')
plt.legend(title='Room type ')
plt.show()
```

how has the no. of reviews change over time 

```python
data['last review']=data['last review']=pd.to_datetime(data['last review'])
review_over_time= data.groupby(data['last review'].dt.to_period('M')).size()
plt.figure(figsize=(8,5))
review_over_time.plot(kind='line',color='blue')
plt.title('Number of Reviews over time ')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
